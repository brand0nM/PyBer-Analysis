```python
import matplotlib.pyplot as plt
```

```python
# percent is the magic function list of other values that can follow
%matplotlib inline
```

```python
savings = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
plt.plot(months, savings)
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Savings Account Results')
plt.show()
```

```python
# Line chart using pythons OOI, Object-Oriented Interface
#create figure with ax.plot

fig, ax = plt.subplots()
# stores a tuple of the objects axis, then it stores these results in a variable called fig and ax
#fig referes to the whole figure while the ax refesrs to the figures axis

ax.plot(months, savings)
#usse ax's plot function to print these results
```
